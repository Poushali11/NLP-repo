```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'become', 'will', 'know', 'use','keep','also','us'])
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
data = df1['VideoEnglishText'].values.tolist()
data_words = list(sent_to_words(data))
```

```python
# Removing the stop words
data_words = list(remove_stopwords(data))
```

```python
import gensim.corpora as corpora
# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
from operator import itemgetter
index_high_corr_list = []

index_counter = 0

for topics in doc_lda:
    highest_feature = max(topics, key=itemgetter(1))[0]
    max_correlation = max(topics, key=itemgetter(1))[1]
    index_high_corr_list.append([index_counter, highest_feature, max_correlation])
    index_counter += 1

index_high_corr_list[:5]
```

```python
import matplotlib.pyplot as plt
```

```python
x = [tup[0] for tup in index_high_corr_list]
y = [tup[2] for tup in index_high_corr_list]

plt.figure(figsize=(20,10))
scatter = plt.plot(x,y, '.')
line = plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), 'r')
plt.title('Highest Correlation of Topics', fontsize=30)
plt.xlabel("Video Content Length Rank Ordered", fontsize=20)
plt.ylabel("Highest Topic Correlation", fontsize=20)
plt.show(scatter, line)
```

```python
def tokenization_report(bow, video_number):
    video_tokens = bow[video_number]
    sorted_tokens = sorted(video_tokens, key=lambda x: x[1], reverse=True)
    for i in range(len(bow[video_number])):
        print("Word {} (\"{}\") appears {} time(s).".format(sorted_tokens[i][0], 
                                                         id2word[sorted_tokens[i][0]], 
                                                         sorted_tokens[i][1]))
tokenization_report(corpus, 0)
```

```python
def video_recommender(title):
    books_checked = 0
    for i in range(len(df1)):
        recommendation_scores = []
        if df1[i][0] == title:
            lda_vectors = doc_lda[i]
            sims = index[lda_vectors]
            sims = list(enumerate(sims))
            for sim in sims:
                video_num = sim[0]
                recommendation_score = [df1[video_num][0], sim[1]]
                recommendation_scores.append(recommendation_score)
            
            recommendation = sorted(recommendation_scores, key=lambda x: x[1], reverse=True)
            print("Your video's most prominent tokens are:")
            video_tokens = corpus[i]
            sorted_tokens = sorted(video_tokens, key=lambda x: x[1], reverse=True)
            sorted_tokens_10 = sorted_tokens[:10]
            for i in range(len(sorted_tokens_10)):
                print("Word {} (\"{}\") appears {} time(s).".format(sorted_tokens_10[i][0], 
                                                             id2word[sorted_tokens_10[i][0]], 
                                                             sorted_tokens_10[i][1]))
            print('-----')
            print("Your video's most prominant topic is:")
            print(lda_model.print_topic(max(lda_vectors, key=lambda item: item[1])[0]))
            print('-----')
            print('Here are your recommendations for "{}":'.format(title))
            display(recommendation[1:11])
        
        else:
            books_checked +=1
            
        if books_checked == len(df1):
            video_suggestions = []
            print('Sorry, but it looks like "{}" is not available.'.format(title))
            other_books = []
            for x in range(len(df1)):
                other_video = df1[x][0]
                video_silimarity = round(cosine_similarity(other_video, title), 2)
                similarity_score = [other_video, video_silimarity]
                video_suggestions.append(similarity_score)

            print('-----')
            ordered_suggestions = sorted(video_suggestions, key=lambda x: x[1], reverse=True)
            print('Were any of the following maybe what you were looking for?')
            print(ordered_suggestions[:10])
```
